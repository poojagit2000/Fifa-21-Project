```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
```

```python
# Reading Training data into a dataframe
Fifa_df = pd.read_csv("fifa21_train.csv")
Fifa_df
```

```python
# List of column values
Fifa_df.columns.values
```

```python
# Datatype of every Column Value
column_data_types = {}
for col in Fifa_df.columns:
    column_data_types[col] = Fifa_df[col].dtype
for col, data_type in column_data_types.items():
    print(f"Column '{col}' has data type: {data_type}")
```

```python
# Decided to drop Unuseful features
col_to_drop = ['ID','Hits','Name','Nationality','Team & Contract','Loan Date End','Joined','Contract','Crossing', 'Finishing', 'Heading Accuracy','Short Passing', 'Volleys','Dribbling', 'Curve','FK Accuracy', 'Long Passing', 'Ball Control','Acceleration', 'Sprint Speed', 'Agility', 'Reactions', 'Balance','Shot Power', 'Jumping', 'Stamina', 'Strength',
       'Long Shots','Aggression', 'Interceptions','Positioning', 'Vision', 'Penalties', 'Composure','Marking', 'Standing Tackle', 'Sliding Tackle','GK Diving', 'GK Handling', 'GK Kicking', 'GK Positioning',
       'GK Reflexes','PAC','SHO','PAS','DRI','DEF','PHY']
Fifa_df.drop(columns = col_to_drop,inplace= True)
Fifa_df.columns.values
```

```python
Fifa_df.shape
```

```python
Fifa_df
```

```python
# Checking the null values
Fifa_df.isnull().sum()
```

```python
# Dropping the Null Values
Fifa_df.dropna(subset= ['Position','Club','A/W'],inplace =True)
Fifa_df.isna().sum()
```

```python
# First, decided to remove '+' sign and Keeping only the value in left for the below list of columns
splitting_col = ['LS', 'ST', 'RS', 'LW', 'LF', 'CF', 'RF', 'RW', 'LAM', 'CAM','RAM', 'LM', 'LCM', 'CM', 'RCM', 'RM', 'LWB', 'LDM', 'CDM', 'RDM','RWB', 'LB', 'LCB', 'CB', 'RCB', 'RB', 'GK']
for col in splitting_col:
    Fifa_df[col] = Fifa_df[col].str.split('+').str[0].str.strip()
    Fifa_df[col] = pd.to_numeric(Fifa_df[col])
Fifa_df
```

```python
# Defining a function, Convert from 'feet-inches' to 'cm'
Fifa_df['Height'] = Fifa_df['Height'].str.replace('"','').astype(str)
Fifa_df['Height'].unique()
def convert_to_cm(h):
    feet,inches = map(int,h.split("'"))
    tot_inc = feet * 12 + inches
    return tot_inc * 2.54
Fifa_df['Height'] = Fifa_df['Height'].apply(convert_to_cm)
Fifa_df
```

```python
# Removing the '€' and replacing 'K' and 'M' with their values 
Change_values = ['Value','Wage','Release Clause']
for col in Change_values:
    Fifa_df[col] = Fifa_df[col].replace({'€': '', ' ': '', 'M': 'e+06', 'K': 'e+03'}, regex=True).astype(float).astype(int)
Fifa_df
```

```python
# Converting Weight from "lbs" to "Kg"
Fifa_df['Weight'] = Fifa_df['Weight'].str.replace("lbs",'').astype(float).astype(int)
Fifa_df['Weight'] = Fifa_df['Weight'] * 0.4536
Fifa_df['Weight']
```

```python
# Removing the "★" Symbol
Fifa_df['SM'] = Fifa_df['SM'].str.replace("★",'').astype(int)
```

```python
remove_star = ['W/F','IR']
for col in remove_star:
    Fifa_df[col] = Fifa_df[col].str.replace(" ★",'').astype(int)
Fifa_df
```

## Defining Clean_df function

```python
def clean_df(fifa_validation):
    # Dropping features
    col_to_drop = ['ID','Hits','Name','Nationality','Team & Contract','Loan Date End','Joined','Contract','Crossing', 'Finishing', 'Heading Accuracy','Short Passing', 'Volleys','Dribbling', 'Curve','FK Accuracy', 'Long Passing', 'Ball Control','Acceleration', 'Sprint Speed', 'Agility', 'Reactions', 'Balance','Shot Power', 'Jumping', 'Stamina', 'Strength',
       'Long Shots','Aggression', 'Interceptions','Positioning', 'Vision', 'Penalties', 'Composure','Marking', 'Standing Tackle', 'Sliding Tackle','GK Diving', 'GK Handling', 'GK Kicking', 'GK Positioning',
       'GK Reflexes','PAC','SHO','PAS','DRI','DEF','PHY']
    fifa_validation.drop(columns = col_to_drop,inplace= True)
    # Dropping null values
    fifa_validation.dropna(subset= ['Position','Club','A/W'],inplace =True)

    
    # First we decided to clean all object columns with '+' sign, and keep the value on the 'left'    
    splitting_col = ['LS', 'ST', 'RS', 'LW', 'LF', 'CF', 'RF', 'RW', 'LAM', 'CAM','RAM', 'LM', 'LCM', 'CM', 'RCM', 'RM', 'LWB', 'LDM', 'CDM', 'RDM','RWB', 'LB', 'LCB', 'CB', 'RCB', 'RB', 'GK']
    for col in splitting_col:
        fifa_validation[col] = fifa_validation[col].str.split('+').str[0].str.strip()
        fifa_validation[col] = pd.to_numeric(fifa_validation[col])
    
    # Cleaning Height and Converting it from feet-inches to cm
    fifa_validation['Height'] = fifa_validation['Height'].str.replace('"','').astype(str)
    fifa_validation['Height'].unique()
    def convert_to_cm(h):
        feet,inches = map(int,h.split("'"))
        tot_inc = feet * 12 + inches
        return tot_inc * 2.54
    fifa_validation['Height'] = fifa_validation['Height'].apply(convert_to_cm)
    
    # Removing € sign, and fitting 'K', 'M', in same scale
    Change_values = ['Value','Wage','Release Clause']
    for col in Change_values:
        fifa_validation[col] = fifa_validation[col].replace({'€': '', ' ': '', 'M': 'e+06', 'K': 'e+03'}, regex=True).astype(float).astype(int)
    
    # Converting From pounds to kilogram (so it is consistent with height conversion in 'cm')
    
    fifa_validation['Weight']=fifa_validation['Weight'].str.replace('lbs','').astype(int)*0.453592
    
    #Cleaning features with '★' sign
    fifa_validation['SM'] = fifa_validation['SM'].str.replace("★",'').astype(int)

    remove_star = ['W/F','IR']
    for col in remove_star:
        fifa_validation[col] = fifa_validation[col].str.replace(" ★",'').astype(int)
    
    return fifa_validation
```
